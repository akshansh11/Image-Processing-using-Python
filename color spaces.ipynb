# **Conversion of RGB color space to HSV color space and vice versa**

```python
from skimage import io
from skimage import color
from skimage import data
from pylab import *
#Read image
img = io.imread('brass microstructure.jpg')
#Convert to HSV
img_hsv = color.rgb2hsv(img)
#Convert back to RGB
img_rgb = color.hsv2rgb(img_hsv)
#Show both figures
figure(0)
io.imshow(img_hsv)
figure(1)
io.imshow(img_rgb)
```
